```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Study_Hours": [2, 3, 1, 4, 5, 6, 3],
    "Productivity": [50, 60, 45, 70, 80, 90, 65]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Pandas Logic
df['Efficiency'] = df["Productivity"] / df["Study_Hours"]
```

```python
# Bubble chart
plt.figure(figsize=(8,6))
plt.scatter(
    df["Study_Hours"],
    df["Productivity"],
    s=df["Efficiency"] * 40,   # bubble size
    alpha=0.6
)
```

```python
# Bubble chart
plt.figure(figsize=(8,6))
plt.scatter(
    df["Study_Hours"],
    df["Productivity"],
    s=df["Efficiency"] * 40,   # bubble size
    alpha=0.6
)

for i, day in enumerate(df["Day"]):
    plt.text(df["Study_Hours"][i]+0.05, df["Productivity"][i]+0.5, day)

plt.xlabel("Study Hours")
plt.ylabel("Productivity")
plt.title("Productivity vs Study Hours (Bubble = Efficiency)")
plt.grid(True)
plt.show()
```
