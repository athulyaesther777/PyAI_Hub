```python
import pygame  # Import the pygame library
import random  # For generating random colors

# Initialize pygame
pygame.init()

# Set screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball Animation")

# Define colors
WHITE = (255, 255, 255)

# Ball properties
ball_radius = 30
ball_x = random.randint(ball_radius, WIDTH - ball_radius)
ball_y = random.randint(ball_radius, HEIGHT - ball_radius)
ball_dx = 5  # Speed in X direction
ball_dy = 5  # Speed in Y direction
ball_color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

# Game loop
running = True
clock = pygame.time.Clock()

while running:
    screen.fill(WHITE)  # Clear screen

    # Move the ball
    ball_x += ball_dx
    ball_y += ball_dy

    # Bounce off walls
    if ball_x - ball_radius <= 0 or ball_x + ball_radius >= WIDTH:
        ball_dx = -ball_dx  # Reverse X direction
        ball_color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))  # Change color

    if ball_y - ball_radius <= 0 or ball_y + ball_radius >= HEIGHT:
        ball_dy = -ball_dy  # Reverse Y direction
        ball_color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))  # Change color

    # Draw the ball
    pygame.draw.circle(screen, ball_color, (ball_x, ball_y), ball_radius)

    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    pygame.display.flip()  # Update display
    clock.tick(60)  # Limit frame rate to 60 FPS

# Quit pygame
pygame.quit()
```

```python
import pygame
import random
import math

# Initialize pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Professional Particle Explosion")

# Colors
BLACK = (0, 0, 0)

# Particle class
class Particle:
    def __init__(self, x, y, color, speed, angle, gravity=0.1, friction=0.98):
        self.x = x
        self.y = y
        self.color = color
        self.speed = speed
        self.angle = angle
        self.gravity = gravity  # Downward force
        self.friction = friction  # Slows down over time
        self.radius = random.randint(3, 6)
        self.alpha = 255  # Opacity
        self.lifetime = random.randint(40, 100)

    def update(self):
        # Apply motion
        self.x += self.speed * math.cos(self.angle)
        self.y += self.speed * math.sin(self.angle) + self.gravity

        # Reduce speed over time (friction effect)
        self.speed *= self.friction

        # Reduce opacity & shrink particle
        self.alpha = max(0, self.alpha - 5)
        self.radius = max(1, self.radius - 0.05)
        self.lifetime -= 1

    def draw(self, surface):
        if self.lifetime > 0:
            particle_surface = pygame.Surface((self.radius * 2, self.radius * 2), pygame.SRCALPHA)
            pygame.draw.circle(particle_surface, (self.color[0], self.color[1], self.color[2], self.alpha), 
                               (self.radius, self.radius), self.radius)
            surface.blit(particle_surface, (self.x - self.radius, self.y - self.radius))

# Generate particles
def create_explosion(x, y):
    particles = []
    for _ in range(80):  # More particles for a grander effect
        speed = random.uniform(3, 7)
        angle = random.uniform(0, 2 * math.pi)
        color = (random.randint(150, 255), random.randint(50, 200), random.randint(150, 255))
        particles.append(Particle(x, y, color, speed, angle))
    return particles

# Main loop
running = True
clock = pygame.time.Clock()
particles = []

while running:
    screen.fill(BLACK)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:  # Click to create explosion
            particles.extend(create_explosion(*event.pos))

    # Update and draw particles
    for particle in particles[:]:
        particle.update()
        particle.draw(screen)
        if particle.lifetime <= 0:
            particles.remove(particle)

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```

```python
import pygame
import random

# Initialize pygame
pygame.init()

# Screen settings
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball Animation")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Font for FPS display
font = pygame.font.Font(None, 24)


class Ball:
    """Class to manage ball properties and behavior."""
    def __init__(self, x, y, radius, dx, dy, color):
        self.x = x
        self.y = y
        self.radius = radius
        self.dx = dx
        self.dy = dy
        self.color = color

    def move(self):
        """Move the ball and handle wall collisions."""
        self.x += self.dx
        self.y += self.dy

        # Collision with walls
        if self.x - self.radius <= 0 or self.x + self.radius >= WIDTH:
            self.dx = -self.dx
            self.change_color()

        if self.y - self.radius <= 0 or self.y + self.radius >= HEIGHT:
            self.dy = -self.dy
            self.change_color()

    def change_color(self):
        """Change the ball's color randomly."""
        self.color = (random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))

    def draw(self, surface):
        """Draw the ball on the given surface."""
        pygame.draw.circle(surface, self.color, (self.x, self.y), self.radius)


def display_fps(clock):
    """Display the current FPS on the screen."""
    fps_text = font.render(f"FPS: {int(clock.get_fps())}", True, BLACK)
    screen.blit(fps_text, (10, 10))


def main():
    """Main game loop."""
    clock = pygame.time.Clock()

    # Create a ball instance with random initial position and color
    ball = Ball(
        x=random.randint(50, WIDTH - 50),
        y=random.randint(50, HEIGHT - 50),
        radius=30,
        dx=5,
        dy=5,
        color=(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255))
    )

    running = True
    while running:
        screen.fill(WHITE)  # Clear screen

        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Update and draw ball
        ball.move()
        ball.draw(screen)

        # Display FPS
        display_fps(clock)

        pygame.display.flip()  # Update screen
        clock.tick(60)  # Maintain 60 FPS

    pygame.quit()


if __name__ == "__main__":
    main()
```

```python
import pygame
import random

# Initialize pygame
pygame.init()

# Screen settings
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Gravity Bouncing Ball with Random Colors")

# Colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Ball properties
GRAVITY = 0.5  # Gravity effect
FRICTION = 0.90  # Energy loss on bounce
TRAIL_LENGTH = 10  # Number of previous positions to store


class Ball:
    """A ball that falls with gravity and bounces with friction and color changes."""
    def __init__(self, x, y, radius):
        self.x = x
        self.y = y
        self.radius = radius
        self.dx = random.uniform(-5, 5)  # Random horizontal speed
        self.dy = random.uniform(-5, 5)  # Random vertical speed
        self.color = self.random_color()
        self.trail = []  # Stores previous positions

    def random_color(self):
        """Generate a random RGB color."""
        return (random.randint(50, 255), random.randint(50, 255), random.randint(50, 255))

    def move(self):
        """Update the ball's position with gravity, bounce mechanics, and color changes."""
        self.dy += GRAVITY  # Apply gravity
        self.x += self.dx
        self.y += self.dy

        # Collision with floor
        if self.y + self.radius >= HEIGHT:
            self.y = HEIGHT - self.radius  # Keep ball above ground
            self.dy = -self.dy * FRICTION  # Reverse direction and reduce speed
            self.color = self.random_color()  # Change color on bounce

        # Collision with walls
        if self.x - self.radius <= 0 or self.x + self.radius >= WIDTH:
            self.dx = -self.dx * FRICTION  # Reverse direction and reduce speed
            self.color = self.random_color()  # Change color on bounce

        # Store position for trail effect
        self.trail.append((self.x, self.y))
        if len(self.trail) > TRAIL_LENGTH:
            self.trail.pop(0)  # Remove oldest position

    def draw(self, surface):
        """Draw the ball and its trail with transparency."""
        for i, (tx, ty) in enumerate(self.trail):
            alpha = int(255 * (i / len(self.trail)))  # Fade effect
            trail_color = (*self.color, alpha)
            pygame.draw.circle(surface, trail_color, (int(tx), int(ty)), self.radius // 2)

        pygame.draw.circle(surface, self.color, (int(self.x), int(self.y)), self.radius)  # Main ball


def main():
    """Main game loop."""
    clock = pygame.time.Clock()
    ball = Ball(WIDTH // 2, HEIGHT // 4, 30)

    running = True
    while running:
        screen.fill(WHITE)  # Clear screen

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        ball.move()
        ball.draw(screen)

        pygame.display.flip()
        clock.tick(60)

    pygame.quit()


if __name__ == "__main__":
    main()
```
