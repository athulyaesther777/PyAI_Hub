```python
# AI Agent Class
class SimpleAIAgent:
  def __init__(self):
    self.memory = {}
    print("AI Agent ü§ñ Online")

  def observe(self, user_input):
    return user_input.lower()

  def decide(self, observation):
    if "my name is" in observation:
      name = observation.replace("my name is", "").strip()
      self.memory["name"] = name # normal to memory

      return f"Nice to meet you, {name}!"

    elif "what is my name" in observation:
      return self.memory.get("name", "I don't know your name yet")

    elif "hello" in observation or "hi" in observation:
      return "Hello! How can I help you?"

    elif "bye" in observation:
      return "Goodbye! See you soon üôã"

    else:
      return "I'm still learning. Can you tell me more?"

  def act(self, response):
    print("ü§ñ AI Agent:", response)

  def run(self):
    while True:
      user_input = input("You: ")
      observation = self.observe(user_input)
      response = self.decide(observation)
      self.act(response)
      if "bye" in observation:
        break
```

```python
# Run the AI agent
agent = SimpleAIAgent()
agent.run()
```
